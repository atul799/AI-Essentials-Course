```python
# Class and Objects
# Define a class called Car
class Car:
    # This method runs when a new Car object is created
    def __init__(self, model, make, engine, style, autonomy=None):
        # Attributes (data) of the car
        self.model = model          # Car model (e.g. BMW iX3)
        self.make = make            # Year of manufacture
        self.engine = engine        # Engine type (EV or ICE Petrol)
        self.style = style          # Body style (SUV, Sedan, etc.)
        self.autonomy = autonomy    # Driving autonomy (optional)

    # Method to switch the car on
    def switch_on(self):
        print("insert key")
        print("turn key or push ignition button")
        print(self.model, "is switched on.")

    # Method to drive the car
    def drive(self):
        print("engage gear")
        print("control steering wheel")
        print(self.model, "is driving.")

    # Method to print all car information
    def info(self):
        print("Model:", self.model)
        print("Make:", self.make)
        print("Engine:", self.engine)
        print("Style:", self.style)

        # Check if autonomy information exists
        if self.autonomy is not None:
            print("Autonomy:", self.autonomy)
        else:
            print("Autonomy: None")

        print("---------------------------")

#Creating Car Objects (Instances)
# Car 1: BMW iX3
car1 = Car(
    model="BMW iX3",
    make=2026,
    engine="EV",
    style="SUV",
    autonomy="Semiautonomous"
)

# Car 2: BMW 5er
car2 = Car(
    model="BMW 5er",
    make=2025,
    engine="ICE Petrol",
    style="Combi",
    autonomy="Assisted"
)

# Car 3: VW Golf (no autonomy attribute)
car3 = Car(
    model="VW Golf",
    make=2023,
    engine="ICE Petrol",
    style="Hatchback"
)

# Car 4: Tesla Model 3
car4 = Car(
    model="Tesla Model 3",
    make=2024,
    engine="EV",
    style="Sedan",
    autonomy="Semiautonomous"
)



#Using the Objects and Methods
# Use methods of each car object
car1.switch_on()
car1.drive()
car1.info()

car2.switch_on()
car2.drive()
car2.info()

car3.switch_on()
car3.drive()
car3.info()

car4.switch_on()
car4.drive()
car4.info()
```

```python
'''Miniâ€‘Project

students notes are available for 5 subjects, he has 30 hours in a week
improve grade in each subject to euqual or below 2.5 and on avg 3
plot the improvement

'''

# Import matplotlib for creating plots
# This library helps us visualize data using charts
import matplotlib.pyplot as plt

#----------Step 1 -------------------------------------
#define data structures

# A list stores multiple items in a fixed order
# We store all subject names here
subjects = ["Maths", "History", "German", "Science", "Ethics"]

# Dictionary stores data as key-value pairs
# Key   â†’ Subject name
# Value â†’ Current note for that subject
current_notes = {
    "Maths": 3.0,
    "History": 3.0,
    "German": 4.0,
    "Science": 3.0,
    "Ethics": 3.0
}

# This dictionary tells us how many study hours are needed
# to improve a subject by 0.5 note
hours_per_half_grade = {
    "Maths": 5,
    "History": 3,
    "German": 5,
    "Science": 3,
    "Ethics": 2
}


#-------------Step2------------------------------------
#define functions

# This function checks whether a subject note
# is higher (worse) than our target note of 2.5
def needs_improvement(note):
    # If the note is 2.5 or higher, improvement is needed
    if note >= 2.5:
        return True
    else:
        return False

# This function calculates how many study hours
# are needed to bring all subjects below note 2.5
def calculate_study_plan(notes, improvement_cost):

    # Dictionary to store how many hours to study per subject
    study_plan = {}

    # Variable to keep track of total required study hours
    total_hours = 0

    # Loop through each subject and its current note
    for subject, note in notes.items():

        # Check if this subject needs improvement
        if needs_improvement(note):

            # Calculate how many 0.5 steps are needed
            # Example: from 4.0 to 2.0 â†’ (4.0 - 2.0) / 0.5 = 4 steps
            steps_needed = int((note - 2.0) / 0.5)

            # Calculate total hours for this subject
            hours_needed = steps_needed * improvement_cost[subject]

            # Store result in study plan
            study_plan[subject] = hours_needed

            # Add to total hours
            total_hours += hours_needed

        else:
            # If no improvement needed, assign 0 hours
            study_plan[subject] = 0

    # Return both the study plan and total hours required
    return study_plan, total_hours

#-------------Step 3----------------------------------------
#Define constraints

# Total number of hours the student can study per week
available_hours = 30

#-----------Step 4-------------------------------------------
#execution of program

# Call the function to calculate study hours
study_plan, total_required_hours = calculate_study_plan(
    current_notes,
    hours_per_half_grade
)

# Print the study plan in a readable way
print(" Weekly Study Plan")
print("-------------------")

for subject, hours in study_plan.items():
    print(subject, "â†’", hours, "hours")

print("\nTotal study hours needed:", total_required_hours)

#check time constraint
# Check if the study plan fits into available time
if total_required_hours <= available_hours:
    print("âœ… Study plan fits within available time.")
else:
    print("âŒ Not enough study hours available.")


#check average note
# Variable to store sum of all notes
total_notes = 0

# Loop through all subject notes
for note in current_notes.values():
    total_notes += note

# Calculate average
average_note = total_notes / len(current_notes)

print("\nCurrent average note:", average_note)

# Check if average goal is met
if average_note <= 3.0:
    print("âœ… Average goal achieved.")
else:
    print("âš  Average needs improvement.")


#----------------Step 5-----------------------------
#Visualization
# Prepare data for plotting
subject_names = list(current_notes.keys())
notes = list(current_notes.values())

# Create a bar chart
plt.bar(subject_names, notes)

# Draw a horizontal line showing target note (2.5)
plt.axhline(y=2.5, linestyle="--")

# Add labels and title
plt.xlabel("Subjects")
plt.ylabel("Current Note (Lower is Better)")
plt.title("Student Current Grades Overview")

# Show the plot
plt.show()
```

```python
'''
Pet Health Tracker program
The pet's hunger level and feed is tracked
for 7 days.
hunger level of pet is generated and then user input is given as feed.
report satus of pet's health each day.
Data is organized and visualized using Pandas and Matplotlib.
'''


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Setup the Simulation
days = 7
pet_name = "Cyber-Hamster"
health_data = []

print(f"--- Starting {days}-Day Health Check for {pet_name} ---")

for day in range(1, days + 1):
    # Use NumPy to 'estimate' hunger (random value between 5 and 10)
    daily_hunger = np.random.randint(5, 11)

    print(f"\nDay {day}: {pet_name}'s hunger level is {daily_hunger}.")

    #try except is a way to provide program a graceful exit for runtime crash
    try:
        # Get user input for food
        food_given = int(input(f"How many treats will you give {pet_name}? "))
    except ValueError:
        print("Invalid input! Giving 0 treats.")
        food_given = 0

    # Logic: Is the pet healthy today?
    is_healthy = food_given >= daily_hunger
    status = "Healthy" if is_healthy else "UNHEALTHY"

    if not is_healthy:
        print(f"ALERT: {pet_name} is {status}! (Need: {daily_hunger} | Got: {food_given})")
    else:
        print(f"Great! {pet_name} is happy and full.")

    # Save data to our list
    health_data.append({
        "Day": day,
        "Hunger": daily_hunger,
        "Food": food_given,
        "Status": status
    })

# 2. Use Pandas to organize the data
df = pd.DataFrame(health_data)
print("\n--- Final Health Report ---")
print(df)

# 3. Use Matplotlib to visualize the trend

plt.figure(figsize=(10, 5))
plt.plot(df["Day"], df["Hunger"], label="Hunger Level", color="red", marker='o')
plt.bar(df["Day"], df["Food"], label="Food Given", color="skyblue", alpha=0.7)

# Add a "Danger Zone" line
plt.axhline(y=df["Hunger"].mean(), color='gray', linestyle='--', label="Avg Hunger")

plt.title(f"{pet_name}'s Health Dashboard")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
