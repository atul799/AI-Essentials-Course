```python
# Class and Objects
# Define a class called Car
class Car:
    # This method runs when a new Car object is created
    def __init__(self, model, make, engine, style, autonomy=None):
        # Attributes (data) of the car
        self.model = model          # Car model (e.g. BMW iX3)
        self.make = make            # Year of manufacture
        self.engine = engine        # Engine type (EV or ICE Petrol)
        self.style = style          # Body style (SUV, Sedan, etc.)
        self.autonomy = autonomy    # Driving autonomy (optional)

    # Method to switch the car on
    def switch_on(self):
        print("insert key")
        print("turn key or push ignition button")
        print(self.model, "is switched on.")

    # Method to drive the car
    def drive(self):
        print("engage gear")
        print("control steering wheel")
        print(self.model, "is driving.")

    # Method to print all car information
    def info(self):
        print("Model:", self.model)
        print("Make:", self.make)
        print("Engine:", self.engine)
        print("Style:", self.style)

        # Check if autonomy information exists
        if self.autonomy is not None:
            print("Autonomy:", self.autonomy)
        else:
            print("Autonomy: None")

        print("---------------------------")

#Creating Car Objects (Instances)
# Car 1: BMW iX3
car1 = Car(
    model="BMW iX3",
    make=2026,
    engine="EV",
    style="SUV",
    autonomy="Semiautonomous"
)

# Car 2: BMW 5er
car2 = Car(
    model="BMW 5er",
    make=2025,
    engine="ICE Petrol",
    style="Combi",
    autonomy="Assisted"
)

# Car 3: VW Golf (no autonomy attribute)
car3 = Car(
    model="VW Golf",
    make=2023,
    engine="ICE Petrol",
    style="Hatchback"
)

# Car 4: Tesla Model 3
car4 = Car(
    model="Tesla Model 3",
    make=2024,
    engine="EV",
    style="Sedan",
    autonomy="Semiautonomous"
)



#Using the Objects and Methods
# Use methods of each car object
car1.switch_on()
car1.drive()
car1.info()

car2.switch_on()
car2.drive()
car2.info()

car3.switch_on()
car3.drive()
car3.info()

car4.switch_on()
car4.drive()
car4.info()
```

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class SmartPet:
    def __init__(self, name):
        self.name = name
        self.history = [] # To store health over time
        self.hunger = 5.0 # Track Hunger level
        self.feed = 5.0 #  Amount fed to Pet
        self.residual_hunger = 0.0 # Residual Hunger
        self.health="Healthy" # Health status of pet

    def calculate_health(self):
        # Using Math to create a non-linear health score
        # Health decreases faster as hunger increases
        #health_check = 100 * math.exp(0.2 * (self.feed - self.hunger))
        self.residual_hunger += (self.feed - self.hunger)
        health_check = self.residual_hunger
        if health_check > 0.20:
            self.health = "Healthy"
        else:
            self.health = "Unhealthy"
        return round(health_check, 2)

    def log_data(self):
        self.history.append(self.calculate_health())

    def show_stats(self):
        # Using Matplotlib to visualize the pet's health trend
        plt.plot(self.history, marker='o', color='green')
        plt.title(f"{self.name}'s Health Journey")
        plt.xlabel("Game Turns")
        plt.ylabel("Health Score")
        plt.show()

# Quick Logic Simulation
pet = SmartPet("Robo-Dog")
print(f"Hi, my name is {pet.name} and i like to be cared for")
for i in range(5):
    #pet.hunger += np.random.uniform(0.5, 1.5) # NumPy for random floats
    pet.hunger = np.random.uniform(0.5, 1.5) # NumPy for random floats
    pet.feed = np.random.uniform(0.5, 1.5) # NumPy for random floats
    print(f"on day: {i+1}, my hunger level is: {pet.hunger} Calories and I am fed {pet.feed} Calories")
    print(f"on day: {i+1}, I am {pet.health}")
    pet.log_data()

print(f"Final Data Table:\n{pd.Series(pet.history, name='Health')}")
pet.show_stats()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Setup the Simulation
days = 7
pet_name = "Cyber-Hamster"
health_data = []

print(f"--- Starting {days}-Day Health Check for {pet_name} ---")

for day in range(1, days + 1):
    # Use NumPy to 'estimate' hunger (random value between 5 and 10)
    daily_hunger = np.random.randint(5, 11)

    print(f"\nDay {day}: {pet_name}'s hunger level is {daily_hunger}.")

    try:
        # Get user input for food
        food_given = int(input(f"How many treats will you give {pet_name}? "))
    except ValueError:
        print("Invalid input! Giving 0 treats.")
        food_given = 0

    # Logic: Is the pet healthy today?
    is_healthy = food_given >= daily_hunger
    status = "Healthy" if is_healthy else "UNHEALTHY"

    if not is_healthy:
        print(f"ALERT: {pet_name} is {status}! (Need: {daily_hunger} | Got: {food_given})")
    else:
        print(f"Great! {pet_name} is happy and full.")

    # Save data to our list
    health_data.append({
        "Day": day,
        "Hunger": daily_hunger,
        "Food": food_given,
        "Status": status
    })

# 2. Use Pandas to organize the data
df = pd.DataFrame(health_data)
print("\n--- Final Health Report ---")
print(df)

# 3. Use Matplotlib to visualize the trend

plt.figure(figsize=(10, 5))
plt.plot(df["Day"], df["Hunger"], label="Hunger Level", color="red", marker='o')
plt.bar(df["Day"], df["Food"], label="Food Given", color="skyblue", alpha=0.7)

# Add a "Danger Zone" line
plt.axhline(y=df["Hunger"].mean(), color='gray', linestyle='--', label="Avg Hunger")

plt.title(f"{pet_name}'s Health Dashboard")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
