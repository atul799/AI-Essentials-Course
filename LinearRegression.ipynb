* We will introduce a concept called **R<sup>2</sup>** which helps finding out which line captures best all the datapoints in the Linear Regression model

* We can find the starting point to build regression model (best fit line) by using mean of x and y.
* We could use the mean of features (age) which is 3 and target (height) which is 79 and plot this point as represented by orange dot in the plot below. Mean of x and y were calculated in the previous code and printed.

* To draw a straight line we need at least 2 defined points, we only have one which is mean of x and mean of y.
* Remember, Equation of line (linear regression model) is y=m * x + c
* m(slope) =(y2-y1)/(x2-x1)
* c(intercept) = value of y when x=0

* with mean as our first point, we can derive slope:
$ m= (x-x)*(y-y)/(x-x)^2 $
* x and y are mean values of x and y respectively
* Intutively, we are using the errors from mean to derive a second point to find a line (slope and y intercept)

* Let's find deviation from mean, we can find how much each of the values in x and y are deviating from mean values (error from mean), this we can do by subtracting each data points from mean value as shown in table below:

|x|y|x-mean_x|y-mean_y|
|--|--|--|--|
|1|50|1-3=-2|50-79=-29|
|2|75|-1|-4|
|3|80|0|1|
|4|80|1|1|
|5|110|2|31|

* Notice, there are negative and positive numbers when you subtract values from mean, positive and negative numbers can cel out error and we dont want that. wW want to capture all of the error, What can we do??
* We can square the difference to include all errors, this helps us in 2 ways:
  * negative values become positive so we can now add all errors
  * if we square error we are scaling the magnitude of error so an error of 10 gives us 100 while error of 20 gives us 400 which means an error of 2x in value (20/10) gives us error os 4x (400/100), hence we are penalizing the higher error more than smaller error!!
* Let's do Square of errors in the table and sum all errors in x and y:

|x|y|x-mean_x|y-mean_y|(x-mean_x)**2|(y-mean_y)**2|(x-mean_x)*(y-mean_y)|
|--|--|--|--|--|--|--|
|1|50|1-3=-2|50-79=-29|4|841|-2*-29=58|
|2|75|-1|-4|1|16|4|
|3|80|0|1|0|1|0|
|4|80|1|1|1|1|1|
|5|110|2|31|4|961|62|
| | | | |Sum=10|Sim=1820|Sum=125|

* Now using above formula we can find the slope:
  * m = 125/10
  * **m = 12.5**
* We still dont have y intercept value.
  * we can find intercept value using mean values in equation of line
  * mean_y = m * mean_x + c : we know values of mean_y,mean_x and slope
  * 79 = 12.5 * 3 + c
  * c = 79 - 37.5
  * **c = 41.5**

* So, We have the line equation passing through the mean.
* y = 12.5 * x + 12.5


* Lets do the steps above in python and plot the line and error from dataset3

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Age (x) and Height (y)
x = np.array([1, 2, 3, 4, 5]) # converrting list into a vector using numpy
y = np.array([50, 75, 80, 80, 110]) # converrting list into a vector

# 1. Calculate Means
x_bar = np.mean(x)
y_bar = np.mean(y)

# 2. Calculate Slope (m) using the formula: sum((x-xbar)*(y-ybar)) / sum((x-xbar)**2)
numerator = np.sum((x - x_bar) * (y - y_bar))
denominator = np.sum((x - x_bar)**2)
m = numerator / denominator

# 3. Calculate Intercept (b) using the mean point (x_bar, y_bar)
# Equation: y = mx + c  =>  c = y - mx
c = y_bar - m * x_bar

# 4. Generate Predictions and Errors
y_pred = m * x + c
errors = y - y_pred

# --- Printing Results ---
print(f"Slope (m): {m:.2f}")
print(f"Intercept (c): {c:.2f}")
print(f"Line Equation: y = {m:.2f}x + {c:.2f}")
for i, err in enumerate(errors):
    print(f"Point {i+1} (x={x[i]}): Error = {err:.2f}")

# --- Plotting ---
plt.figure(figsize=(10, 6)) #defining size of plot
#actual data point plot
plt.scatter(x, y, color='blue', label='Actual Data', zorder=5)
#predicted line plot
plt.plot(x, y_pred, color='red', label='Regression Line', linewidth=2)

# Predicted Points on the Line
plt.scatter(x, y_pred, color='red', marker='o', facecolors='none', s=100, label='Predicted Points $(\hat{y})$', zorder=5)

# Annotate Errors (Residuals)
for i in range(len(x)):
    # Draw vertical line representing the error
    plt.vlines(x[i], ymin=y[i], ymax=y_pred[i], colors='gray', linestyles='dotted')
    # Label the error value
    plt.text(x[i] + 0.1, (y[i] + y_pred[i])/2, f'err: {errors[i]:.1f}',
             fontsize=9, color='purple', fontweight='bold')

plt.xlabel('Age (Years)')
plt.ylabel('Height (cm)')
plt.title('Linear Regression: Age vs Height with Error Annotations')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

* We now have a prediction model, is it the best Linear regression model (line) ?
* Remember our goal is to get a model which fits the given dataset best, meaning least(smallest) error between predicted and actual points.
* How do we know if the Linear Regression line we got is the best?
* We use R measure to determine best fit. It tells use how well the actual points are predicted by Regression model line.
* Formally: R-squared (R) is a statistical measure, also called the coefficient of determination, that shows how much of the variation in the dependent variable is explained by the independent variable(s) in a regression model, indicating how well the model fits the data (0 to 1, with higher being better fit)
  * If R is 0.75, it means 75% of the variation in the dependent variable is explained by the independent variables in the model
* R is calcuated as ratio of unexplained variation and total variation : variation here is simply sum of errors

$$ R = 1 - (Sum-of-residuals)/(Sum-of-mean-error)   $$

$$ R = 1 - \frac{(y-ypredicted)^2}{(y-ymean)^2} $$

* Lets Calculate R using the dataset3 table:

|x|y|y-mean_y|(y-mean_y)**2|ypredicted|(y-ypredicted**2)|
|--|--|--|--|--|--|
|1|50|50-79=-29|(-29)*(-29)=841|54|(50-54)*(-4)=16|
|2|75|75-79=-4|16|83.5|8.5*8.5=72.25|
|3|80|80-79=1|1|79|1*1=1|
|4|80|80-79=1|1|91.5|-11.5*-11.5=132.25|
|5|110|110-79=31|961|104|6*6=36|
| | | | Sum=1820| |Sum=273.75|

$$ R = 1 - \frac{273.75}{1820} = 0.8585 $$
* An $R^2$ of 0.8585 indicates that approximately 85.85% of the variation in height can be explained by the age of the children in this model, which signifies a very strong linear relationship.
* Lets calculate R using python



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([50, 75, 80, 80, 110])

# 1. Slope and Intercept
x_bar, y_bar = np.mean(x), np.mean(y)
m = np.sum((x - x_bar) * (y - y_bar)) / np.sum((x - x_bar)**2)
c = y_bar - m * x_bar

# 2. Predictions and R2
y_pred = m * x + c
ss_res = np.sum((y - y_pred)**2)
ss_tot = np.sum((y - y_bar)**2)
r2 = 1 - (ss_res / ss_tot)

print(f"R-squared: {r2:.4f}")

# 3. Visualization
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.scatter(x, y_pred, color='red', marker='o', facecolors='none', s=100, label='Predicted Points')

for i in range(len(x)):
    plt.vlines(x[i], ymin=y[i], ymax=y_pred[i], colors='gray', linestyles='dotted')
    plt.text(x[i], y_pred[i]-5, f'({x[i]}, {y_pred[i]:.1f})', color='red', ha='center')

plt.legend()
plt.show()
```

* Below is the python code - prediction values are marked in green squares in the plot:

```python
import numpy as np
import matplotlib.pyplot as plt
#import Linearregression class from sklearn module
from sklearn.linear_model import LinearRegression

# 1. Prepare Data
# sklearn expects data in a specififc format:
# number of columns in matrix shall be equal to number of features in dataset
# in our case we have only 1 feature (age), so reshape (-1,1) sets one column
# and as many rows as there are data point
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([50, 75, 80, 80, 110])

# 2. Fit the Model
# create an object of the LinearRegression class
model = LinearRegression()
# apply the fit method of the LinearRegression class to train the model
model.fit(x, y)

# 3. Predict Height for 6 and 7
#input new data in row,column format
x_new = np.array([[6], [7]])
#predict height using the model that was trained
y_new_pred = model.predict(x_new)

# 4. Visualization
plt.scatter(x, y, color='blue', label='Actual Data (Age 1-5)')
plt.plot(np.append(x, x_new), model.predict(np.append(x, x_new).reshape(-1,1)),
         color='red', linestyle='--', label='Regression Line')

# Highlight new predictions
plt.scatter(x_new, y_new_pred, color='green', marker='s', s=100, label='Predictions (Age 6, 7)')
for i, txt in enumerate(y_new_pred):
    plt.annotate(f"{txt:.1f}", (x_new[i], y_new_pred[i]+2), color='green', weight='bold')

plt.xlabel('Age (Years)')
plt.ylabel('Height (cm)')
plt.title('Sklearn Linear Regression: Prediction for Age 6 and 7')
plt.legend()
plt.show()
```

**Train-Test Split**
* During training of the model we use the dataset including features and target output. Sometimes the model may "memorise" the dataset so while it shows good quality model (R for Linear Regression), it may not work as well for new feature data while doing inference (prediction). This behavior is called **overfitting**
* To avoid overfitting the data is split into Training and Test set during training process, the model is trained with Training set and  checked for quality with the Test set, if quality of model matches between training and test set only then training is stopped and used for inference (prediction).
* There is a function available in skelarn library which does train-test split, so you dont have to worry about writing code for this.

```python
# create features array (x) and target output vector (y)
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
y
```

```python
#plots showing correlation between few features and target output
#hours studied to performance
sns.regplot(x=df['Hours Studied'],y=y,color='c')
plt.title('Hours Studied vs Performance')

```

```python
#previous score to performance
sns.regplot(x=df['Previous Scores'],y=y,color='g')
plt.title('Previous Score vs Performance')

```

```python
#sleep hours to Performance
sns.regplot(x=df['Sleep Hours'],y=y,color='m')
plt.title('Sleep Hours vs Performance')

```

```python
#categorical data cant be used directly in a Linear regression they have to be encoded (1-hot encoding)
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
#one hot encoding the categorical feature Extracurricular Activities
col=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),["Extracurricular Activities"]),remainder='passthrough')
x=col.fit_transform(x)
x
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
#Train the Linearregression model and use the model on test set
from sklearn.linear_model import LinearRegression
#train on train dataset
model=LinearRegression()
model.fit(x_train,y_train)

#R2 of trained model
score=model.score(x_train,y_train)
print(f"R2 score of training dataset: {score
                }")

#coefficient of fatures in LinearRegression eqution
#model.coef_
#model intercept
#model.intercept_


#predict test dataset
y_pred=model.predict(x_test)
y_pred
```

```python
#generate difference between actual and predicted value in test dataset
df2=pd.DataFrame({'Actual_Value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_squared_error

#mse of test dataset
MSE=mean_squared_error(y_test,y_pred)
print(f"MSE of test dataset: {MSE}")
```

```python
#R2 score of test dataset--> compare it to training dataset
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(f"R2 score of test dataset: {score}")

```
