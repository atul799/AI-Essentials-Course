```python
import matplotlib.pyplot as plt

# --- 1. FUNCTIONS ---
def analyze_grades(name, score_list):
    avg = sum(score_list) / len(score_list)
    status = "University Ready" if avg > 80 else "Needs Improvement"

    # Create a nice summary string
    summary = f"GRADE REPORT\n"
    summary += f"Student: {name}\n"
    summary += f"Average Score: {avg}\n"
    summary += f"Final Decision: {status}\n"
    return avg, summary

# --- 2. INPUT ---
print("--- Grade Predictor System ---")
user_name = input("Enter student name, then hit Enter: ")
# We ask for scores separated by spaces, then turn them into integers
raw_scores = input("Enter 5 scores separated by spaces (e.g., 80 90 75 88 92): ")
#user input is treated as text(string), however to do math it has to be converted to integer or real number
#in the code below all elements of raw_score are converted to integer (int(s)) by looping through it and saving result
#in new list score_list, the looping method used here is very special syntax of python called syntactic sugar
score_list = [int(s) for s in raw_scores.split()]

# --- 3. PROCESSING ---
#function returns 2 values which are stored in two variables average and final_report
average, final_report = analyze_grades(user_name, score_list) #calling function to analyze grade

# --- 4. OUTPUT (To Screen) ---
print("\n" + final_report) # the + here is simply adding two strings in this case first string is actually a new line char

# --- 5. WRITING TO FILE (Permanent Storage) ---
file_name = f"{user_name}_report.txt"
with open(file_name, "w") as f:
    f.write(final_report)
print(f"Success! A permanent copy has been saved to {file_name}")

# --- 6. PLOTTING (Visualization) ---
plt.bar(range(1, len(score_list)+1), score_list, color='skyblue')
plt.axhline(y=80, color='r', linestyle='--', label='Threshold')
plt.title(f"Score Distribution for {user_name}")
plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.legend()
plt.show()
```
